```python
import numpy as np 
from matplotlib import pyplot as plt 
from PIL import Image
```

```python
# Άνοιγμα εικόνας από τον δίσκο
A = np.array(Image.open('glasses2.png'))
print('Μέγεθος εικόνας. Γραμμές: {}, Στήλες: {}, Κανάλια: {}'.format(
    A.shape[0],
    A.shape[1],
    A.shape[2],
))
```

```python
plt.imshow(A)
plt.show()
```

```python
# Η έγχρωμη εικόνα αναπαρίσταται σαν 3 ξεχωριστοί πίνακες.
# Η μία εικόνα για το κόκκινο, η άλλη για το πράσινο, η τελευταία για το μπλε.
A0 = A[:, :, 0]
A1 = A[:, :, 1]
A2 = A[:, :, 2]
print(A0.shape)
```

```python
# Η πρώτη εικόνα αντιστοιχεί στις εντάσεις του Κόκκινου.
# Παρατηρήστε π.χ. την ένταση του κόκκινου χρώματος στο αριστερό ποτήρι
plt.imshow(A0, cmap='gray')
plt.show()
# Η δεύτερη εικόνα αντιστοιχεί στις εντάσεις του Πράσινου.
# Παρατηρήστε π.χ. την ένταση του πράσινου χρώματος στο τασάκι
plt.imshow(A1, cmap='gray')
plt.show()
# Η τρίτη εικόνα αντιστοιχεί στις εντάσεις του Μπλε.
# Παρατηρήστε π.χ. την ένταση του μπλε χρώματος στα ζάρια
plt.imshow(A2, cmap='gray')
plt.show()
```

```python
# Ψευδοχρωματισμός
# Μπορούμε από την python να επιλέξουμε διαφορετικές παλέττες για να ψευδοχρωματίσουμε
# μια οποιαδήποτε grayscale εικόνα.
A = np.array(Image.open('thermal-ir-comparison-2.jpg'))
print(A.shape)
# Αλλάζοντας την παράμετρο 'cmap', μπορώ να πετύχω ψευδοχρωματισμό με μια μεγάλη επιλογή από παλέττες
# https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
plt.imshow(A, cmap='coolwarm')
plt.show()
```
